## Mean-Variance-Standard Deviation Calculator

```python
import numpy as np

def calculate(lst):
     
    # Ensure the input list contains exactly nine numbers
    
    if len(lst) != 9:
        raise ValueError("List Must Contain Nine Numbers")

    # reshaping the list into 3 by 3 matrix    
    arr = np.array(lst).reshape(3,3)

    # calculating the mean, variance, standard deviation, max, min, sum along rows and columns and flattened matrix
    mean = [list(np.mean(arr, axis = 0)), list(np.mean(arr, axis = 1)), float(np.mean(arr.flatten()))]
    variance = [list(np.var(arr, axis = 0)), list(np.var(arr, axis = 1)), float(np.var(arr.flatten()))]
    std_deviation = [list(np.std(arr, axis = 0)), list(np.std(arr, axis = 1)), float(np.var(arr.flatten()))]
    max_val = [list(np.max(arr, axis = 0)), list(np.max(arr, axis = 1)), float(np.max(arr.flatten()))]
    min_val = [list(np.min(arr, axis = 0)), list(np.min(arr, axis = 1)), float(np.min(arr.flatten()))]
    sum_val = [list(np.sum(arr, axis = 0)), list(np.sum(arr, axis = 1)), float(np.sum(arr.flatten()))]

    # construct the dictionary
    calculations = {
            'mean' : mean,
            'variance' : variance,
            'standard_deviation' : std_deviation,
            'max' : max_val,
            'min' : min_val,
            'sum' : sum_val
        }
        
    return calculations
input_list = [0, 1, 2, 3, 4, 5, 6, 7, 8]

result = calculate(input_list)
print(result)
```

```python
# importing necessary packages

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# reading the dataset and storing in a new variable

medical_data = pd.read_csv('medical_examination.csv')
```

```python
# viewing the dataset

medical_data
```

```python
# creating a copy of the dataset

medical_exam = medical_data.copy()
```

```python
# viewing the first five rows of the copy dataset

medical_data.head()
```

```python
# viewing the first five rows of the original dataset


medical_exam.head()
```

```python
# identifying the columns of the dataset

medical_exam.columns
```

```python
# creating a new column 'overweight' for the dataset

medical_exam['overweight'] = (medical_exam['weight'] / (medical_exam['height'] ** 2) > 25).astype(int)
```

```python
# viewing the result of the dataset with the new column

medical_exam.head()
```

```python
# normalizing the 'cholesterol' column with the specified condition 

medical_exam['cholesterol'] = (medical_exam['cholesterol'] > 1).astype(int)
```

```python
# viewing the results

medical_exam['cholesterol']
```

```python
# normalizing the 'gluc' column with the specified condition 


medical_exam['gluc'] = (medical_exam['gluc'] > 1).astype(int)
```

```python
# viewing the results


medical_exam['gluc']
```

```python
# viewing the final result of the dataset after wrangling the data

medical_exam.head()
```

```python
# defining a function

def draw_cat_plot():
    
# Using pd.melt() to reshape the DataFrame from wide to long format.
# id_vars='cardio': Keeping the 'cardio' column fixed and not melting it.
# value_vars=['cholesterol', 'gluc', 'smoke', 'alco', 'active', 'overweight']: Melting these columns, and their values will become rows under a new column named 'feature'.
# var_name='feature': Specifying the name of the new column where the variable names will be stored.
# value_name='value': Specifying the name of the new column where the values of the melted columns will be stored.
    
    df_cat = pd.melt(medical_exam, id_vars='cardio', value_vars=['cholesterol', 'gluc', 'smoke', 'alco', 'active', 'overweight'], var_name='feature', value_name='value')
    
# Grouping the DataFrame df_cat by three columns: 'feature', 'value', and 'cardio'.
# The .size() function counts the number of occurrences of each group.
# .reset_index(name='total_count') resets the index of the DataFrame and renames the count column to 'total_count'. 
    
    df_cat = df_cat.groupby(['feature', 'value', 'cardio']).size().reset_index(name='total_count')
    
# Creating a categorical plot using Seaborn.
# data=df_cat specifies the DataFrame to use for plotting.
# kind='bar' specifies the type of plot (in this case, a bar plot).
# x='feature' specifies the column to use for the x-axis.
# y='total_count' specifies the column to use for the y-axis.
# hue='value' specifies the column to use for color differentiation.
# col='cardio' specifies the column to use for creating separate plots for each unique value.
    
    g = sns.catplot(data=df_cat, kind='bar', x='feature', y='total_count', hue='value', col='cardio')
    
# Saving the generated figure as a PNG file named 'catplot.png'.

    fig = g.fig
    fig.savefig('catplot.png')
    
# Returning the generated figure.
    
    return fig

# calling the function

draw_cat_plot()
  
```

## Page View Time Series Visualizer

```python
# reading the dataset with the read_csv() method

freecode_date = pd.read_csv('fcc-forum-pageviews.csv', index_col ='date')
```

```python
# viewing the dataset to have an idea of its content

freecode_date
```

```python
# Filtering the 'value' column to include only values within the 2.5th and 97.5th percentiles


freecode_date['value'].between(freecode_date['value'].quantile(0.025), freecode_date['value'].quantile(0.975))
```

```python
# Filtering the dataset to display only the rows where the condition is met


freecode_date[freecode_date['value'].between(freecode_date['value'].quantile(0.025), freecode_date['value'].quantile(0.975))]
```

```python
# converting the date index to a datetime series

freecode_date.index = pd.to_datetime(freecode_date.index)
```

```python
# Plotting the daily freeCodeCamp forum page views from May 2016 to December 2019
# Using the date index converted to a numpy array for x-axis and the 'value' column also converted to a numpy array for y-axis

plt.figure(figsize=(10, 6))
plt.plot(freecode_date.index.to_numpy(), freecode_date['value'].to_numpy(), color='red')

plt.title('Daily freeCodeCamp Forum Page Views 5/2016-12/2019')
plt.xlabel('Date')
plt.ylabel('Page Views')

plt.savefig('line_plot.png')
plt.show()
```

```python
def draw_bar_plot():
    
        # Make a copy of the DataFrame containing the page view data

    
    freecode_camp = freecode_date.copy()
    
        # Extract the year and month from the index and create a new 'year' and 'month' column

    freecode_camp['year'] = freecode_camp.index.year
    freecode_camp['month'] = freecode_camp.index.month_name()
    
        # Group the data by year and month, calculate the mean page views, and reset the index

    freecode_camp = freecode_camp.groupby(['year', 'month'])['value'].mean().reset_index()
    
        # Create a bar plot using matplotlib and seaborn, with separate bars for each month within each year

    plt.figure(figsize = (10,6))
    sns.barplot(data = freecode_camp, x = 'year', y = 'value', hue='month', hue_order=[
                'January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December'])

    plt.xlabel('Years')
    plt.ylabel('Average Page Views')
    plt.title('Average Daily Page Views by Year and Month')
    plt.legend(title = 'month', loc = 'upper left')
    
    plt.savefig('bar_plot.png')
    plt.show()
    
draw_bar_plot()
```

```python
def draw_box_plot():
    freecode_camp_date = freecode_date.copy()
    freecode_camp_date.reset_index(inplace=True)
    
        # Extract the year from the date column and create a new 'year' column

    freecode_camp_date['year'] = [d.year for d in freecode_camp_date.date]
    
        # Extract the month abbreviation from the date column and create a new 'month' column

    freecode_camp_date['month'] = [d.strftime('%b') for d in freecode_camp_date.date]

    
        # Create a subplot with 1 row and 2 columns, specifying the figure size

    plot, axes = plt.subplot(1, 2, figsize = (15,7))
    
        # Create a box plot for the page views grouped by year

    sns.boxplot(data = freecode_camp_date, x = 'year', y = 'value', ax = axes[0])
    axes[0].set_title('Year-wise Box Plot (Trend)')
    axes[0].set_xlabel('Year')
    axes[0].set_ylabel('Page Views')
    
        # Create a box plot for the page views grouped by month, specifying the order of months

    sns.boxplot(data - freecode_camp_date, x = 'month', y = 'value', order = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'], ax = axes[1])
    axes[1].set_title('Month-wise Box Plot (Seasonality)')
    axes[1].set_xlabel('Month')
    axes[1].set_ylabel('Page Views')
    
    plt.savefig('box_plot.png')
    plt.show()
    
draw_bar_plot()  
```

## Sea Level Predictor

```python
# import the necessary packages

import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import linregress
```

```python
# reading the dataset

sea_data = pd.read_csv('epa-sea-level.csv')
```

```python
# printing out the dataset to have an idea of what it looks like

sea_data
```

```python
# plotting a scatter plot with matplotlib to see the relationship with 'Year' and 'CSIRO Adjusted Sea Level'

plt.figure(figsize=(10,6))
plt.scatter(x = 'Year', y = 'CSIRO Adjusted Sea Level', data = sea_data)
plt.xlabel('Year')
plt.ylabel('CSIRO Adjusted Sea Level')
plt.title('CSIRO Adjusted Sea Level Over Time')
plt.legend()
plt.show()
```

```python
# creating a copy of the dataset

sea_level = sea_data.copy()
```

```python
# Extracting the 'Year' and 'CSIRO Adjusted Sea Level' columns and converting them to numpy arrays

x = sea_level['Year'].to_numpy()
y = sea_level['CSIRO Adjusted Sea Level'].to_numpy()

# Creating a new figure with a specific size

plt.figure(figsize=(10,6))

# Performing linear regression on the sea level data
# linregress returns slope, intercept, std_err, r-value, p-value

slope, intercept, std_err, r, p = linregress(x,y)

# Plotting the line of best fit using the calculated slope and intercept

plt.plot(x, slope * x + intercept)

# Predicting the sea level in 2050 based on the linear regression model

predicted_sea_level_2050 = slope * 2050 + intercept

# Plotting a point representing the predicted sea level in 2050

plt.scatter(2050, predicted_sea_level_2050)
```

```python
# filering the dataset to get the records from year 2000 onwards

sea_level[sea_level['Year'] >= 2000]
```

```python
# storing the result into a new variable

sea_level_recent = sea_level[sea_level['Year'] >= 2000]
```

```python
# Extracting the 'Year' and 'CSIRO Adjusted Sea Level' columns and converting them to numpy arrays

x_recent = sea_level_recent['Year'].to_numpy()
y_recent = sea_level_recent['CSIRO Adjusted Sea Level'].to_numpy()

# Performing linear regression on the sea level data
# linregress returns slope, intercept, std_err, r-value, p-value

slope_recent, intercept_recent, r_value_recent, p_value_recent, std_err_recent = linregress(x_recent, y_recent)

# Creating a new figure with a specific size

plt.figure(figsize = (10,6))

# plotting a scatterplot between x_recent and y_recent

plt.scatter(x_recent, y_recent)

# Plotting the line of best fit using the calculated slope and intercept

plt.plot(x_recent, slope_recent * x_recent + intercept_recent)


# Predicting sea level rise in 2050 based on recent trend
predicted_sea_level_2050_recent = slope_recent * 2050 + intercept_recent
plt.scatter(2050, predicted_sea_level_2050_recent, color='red', label='Predicted Sea Level in 2050 (2000 onwards)')

# Adding labels and title
plt.xlabel('Year')
plt.ylabel('CSIRO Adjusted Sea Level')
plt.title('CSIRO Adjusted Sea Level Over Time (2000 onwards)')

# Displaying the plot
plt.grid(True)
plt.legend()
plt.show()
```
